```python
from google.colab import files
uploaded = files.upload()  # sube train.csv primero
uploaded = files.upload()  # luego test.csv

import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print("\n" + "=" * 60)
print("3. ESTADÍSTICAS BÁSICAS")
print("=" * 60)

print("TRAIN")
print(train.describe(include='all').T.to_string())
print("TEST")
print(test.describe(include='all').T.to_string())

print("\n--- Comparación de variables numéricas (medias) ---")
num_cols = ['Age', 'SibSp', 'Parch', 'Fare']
comp = pd.DataFrame({
    'Train_media': train[num_cols].mean(),
    'Test_media' : test[num_cols].mean(),
})
comp['Diferencia_%'] = ((comp['Train_media'] - comp['Test_media']) / comp['Train_media'] * 100).round(2)
print(comp.round(4))

print("""
Comentario: Las medias de Age, SibSp y Parch son muy similares entre
train y test, lo que sugiere que ambos conjuntos provienen de la misma
distribución. La tarifa (Fare) presenta una diferencia pequeña.
""")
```

```python
print("=" * 60)
print("4. VARIABLE FAMILIARES (SibSp + Parch)")
print("=" * 60)

train['Familiares'] = train['SibSp'] + train['Parch']
test['Familiares']  = test['SibSp']  + test['Parch']

print("\nDistribución de 'Familiares' en train:")
print(train['Familiares'].value_counts().sort_index())

print("""
Opinión: La variable Familiares resume el tamaño del grupo familiar de cada
pasajero. La mayoría viajaba solo (Familiares=0). Grupos muy grandes
(≥5) son raros. Esta variable puede ser útil para predecir supervivencia,
ya que familias pequeñas pueden haberse ayudado mutuamente, mientras que
familias grandes tuvieron más dificultad para evacuarse juntas.
""")
```

```python
print("=" * 60)
print("5. UNIÓN VERTICAL")
print("=" * 60)

train['source'] = 'train'
test['source']  = 'test'

# test no tiene 'Survived' → se agrega como NaN
combined = pd.concat([train, test], ignore_index=True, sort=False)

print(f"\nDimensiones del dataset combinado: {combined.shape}")
print(f"Valores en 'source': {combined['source'].value_counts().to_dict()}")
print(f"Valores faltantes en 'Survived': {combined['Survived'].isna().sum()}")
print("""
Nota: La columna 'Survived' en las filas del conjunto test queda como NaN
porque esa información no está disponible (es lo que el modelo debe predecir).
Al analizar supervivencia solo usaremos las filas donde source='train'.
""")
```

```python
print("=" * 60)
print("6. REPRESENTATIVIDAD DE TRAIN Y TEST")
print("=" * 60)

for col in ['Age', 'Fare', 'Pclass', 'Sex', 'Embarked']:
    print(f"\n  {col}:")
    if combined[col].dtype == 'object' or combined[col].nunique() <= 5:
        ct = pd.crosstab(combined[col], combined['source'], normalize='columns').round(3)
        print(ct)
        # --- NUEVA LÍNEA PARA EXPORTAR ---
        ct.to_csv(f'outputs/tablas/representatividad_{col}.csv', index=True)
    else:
        grp = combined.groupby('source')[col].agg(['mean','std']).round(3)
        print(grp)
        # --- NUEVA LÍNEA PARA EXPORTAR ---
        grp.to_csv(f'outputs/tablas/representatividad_{col}.csv', index=True)

print("""
Conclusión: Las distribuciones de Pclass, Sex, Age y Embarked son muy
similares entre train y test, lo que indica que ambos son representativos
de la misma población (el pasaje del Titanic). No hay evidencia de sesgo
sistemático entre conjuntos.
""")
```

```python
print("=" * 60)
print("7. DATOS FALTANTES")
print("=" * 60)

# Cálculo de valores nulos y porcentajes
missing = combined.isnull().sum()
missing_pct = (missing / len(combined) * 100).round(2)
miss_df = pd.DataFrame({'Faltantes': missing, 'Porcentaje %': missing_pct})

# Filtrar solo las columnas que tienen datos faltantes
miss_df = miss_df[miss_df['Faltantes'] > 0].sort_values('Porcentaje %', ascending=False)
print(miss_df)

# Exporta la tabla de datos faltantes a la carpeta exacta
miss_df.to_csv('outputs/tablas/tabla_datos_faltantes.csv', index=True)

print("""
Significado de los datos faltantes:
  - Age (≈20%): La edad no siempre era registrada en los manifiestos.
    Puede imputarse con la mediana por sexo/clase.
  - Cabin (≈77%): Solo los pasajeros de primera clase tenían cabinas
    asignadas. Los de tercera dormían en dormitorios comunes, de ahí
    el alto porcentaje de datos faltantes.
  - Embarked (2 filas): Posiblemente errores de registro. Se puede
    imputar con la moda (Southampton).
  - Fare (1 fila en test): Puede imputarse con la mediana de su clase.
  - Survived: Corresponde a las filas del conjunto test (esperado).
""")
```
