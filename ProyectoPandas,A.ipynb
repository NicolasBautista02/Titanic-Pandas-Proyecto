```python
import pandas as pd

# Carga los archivos directamente desde la carpeta data/ del proyecto
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

print("¡Archivos cargados con éxito desde la carpeta data/!")
```

```python
print("\n" + "=" * 60)
print("3. ESTADÍSTICAS BÁSICAS")
print("=" * 60)

print("TRAIN")
print(train.describe(include='all').T.to_string())
print("TEST")
print(test.describe(include='all').T.to_string())

print("\n--- Comparación de variables numéricas (medias) ---")
num_cols = ['Age', 'SibSp', 'Parch', 'Fare']
comp = pd.DataFrame({
    'Train_media': train[num_cols].mean(),
    'Test_media' : test[num_cols].mean(),
})
comp['Diferencia_%'] = ((comp['Train_media'] - comp['Test_media']) / comp['Train_media'] * 100).round(2)
print(comp.round(4))

print("""
Comentario: Las medias de Age, SibSp y Parch son muy similares entre
train y test, lo que sugiere que ambos conjuntos provienen de la misma
distribución. La tarifa (Fare) presenta una diferencia pequeña.
""")
```

```python
print("=" * 60)
print("4. VARIABLE FAMILIARES (SibSp + Parch)")
print("=" * 60)

train['Familiares'] = train['SibSp'] + train['Parch']
test['Familiares']  = test['SibSp']  + test['Parch']

print("\nDistribución de 'Familiares' en train:")
print(train['Familiares'].value_counts().sort_index())

print("""
Opinión: La variable Familiares resume el tamaño del grupo familiar de cada
pasajero. La mayoría viajaba solo (Familiares=0). Grupos muy grandes
(≥5) son raros. Esta variable puede ser útil para predecir supervivencia,
ya que familias pequeñas pueden haberse ayudado mutuamente, mientras que
familias grandes tuvieron más dificultad para evacuarse juntas.
""")
```

```python
print("=" * 60)
print("5. UNIÓN VERTICAL")
print("=" * 60)

train['source'] = 'train'
test['source']  = 'test'

# test no tiene 'Survived' → se agrega como NaN
combined = pd.concat([train, test], ignore_index=True, sort=False)

print(f"\nDimensiones del dataset combinado: {combined.shape}")
print(f"Valores en 'source': {combined['source'].value_counts().to_dict()}")
print(f"Valores faltantes en 'Survived': {combined['Survived'].isna().sum()}")
print("""
Nota: La columna 'Survived' en las filas del conjunto test queda como NaN
porque esa información no está disponible (es lo que el modelo debe predecir).
Al analizar supervivencia solo usaremos las filas donde source='train'.
""")
```

```python
print("=" * 60)
print("6. REPRESENTATIVIDAD DE TRAIN Y TEST")
print("=" * 60)

for col in ['Age', 'Fare', 'Pclass', 'Sex', 'Embarked']:
    print(f"\n  {col}:")
    if combined[col].dtype == 'object' or combined[col].nunique() <= 5:
        ct = pd.crosstab(combined[col], combined['source'], normalize='columns').round(3)
        print(ct)
        # --- NUEVA LÍNEA PARA EXPORTAR ---
        ct.to_csv(f'outputs/tablas/representatividad_{col}.csv', index=True)
    else:
        grp = combined.groupby('source')[col].agg(['mean','std']).round(3)
        print(grp)
        # --- NUEVA LÍNEA PARA EXPORTAR ---
        grp.to_csv(f'outputs/tablas/representatividad_{col}.csv', index=True)

print("""
Conclusión: Las distribuciones de Pclass, Sex, Age y Embarked son muy
similares entre train y test, lo que indica que ambos son representativos
de la misma población (el pasaje del Titanic). No hay evidencia de sesgo
sistemático entre conjuntos.
""")
```

```python
print("=" * 60)
print("7. DATOS FALTANTES")
print("=" * 60)

# Cálculo de valores nulos y porcentajes
missing = combined.isnull().sum()
missing_pct = (missing / len(combined) * 100).round(2)
miss_df = pd.DataFrame({'Faltantes': missing, 'Porcentaje %': missing_pct})

# Filtrar solo las columnas que tienen datos faltantes
miss_df = miss_df[miss_df['Faltantes'] > 0].sort_values('Porcentaje %', ascending=False)
print(miss_df)

# Exporta la tabla de datos faltantes a la carpeta exacta
miss_df.to_csv('outputs/tablas/tabla_datos_faltantes.csv', index=True)

print("""
Significado de los datos faltantes:
  - Age (≈20%): La edad no siempre era registrada en los manifiestos.
    Puede imputarse con la mediana por sexo/clase.
  - Cabin (≈77%): Solo los pasajeros de primera clase tenían cabinas
    asignadas. Los de tercera dormían en dormitorios comunes, de ahí
    el alto porcentaje de datos faltantes.
  - Embarked (2 filas): Posiblemente errores de registro. Se puede
    imputar con la moda (Southampton).
  - Fare (1 fila en test): Puede imputarse con la mediana de su clase.
  - Survived: Corresponde a las filas del conjunto test (esperado).
""")
```

```python
print("=" * 60)
print("8. PREGUNTAS DESCRIPTIVAS (sobre conjunto combinado / train)")
print("=" * 60)

# a) Edad promedio
edad_prom = combined['Age'].mean()
print(f"\na) Edad promedio de los pasajeros del Titanic: {edad_prom:.2f} años")

# b) Sobrevivientes / fallecidos
survived_counts = train['Survived'].value_counts()
print(f"\nb) Sobrevivientes: {survived_counts[1]}  |  Fallecidos: {survived_counts[0]}")

# c) Tarifa promedio primera clase
fare_1st = train[train['Pclass'] == 1]['Fare'].mean()
print(f"\nc) Tarifa promedio primera clase: ${fare_1st:.2f}")

# d) Pasajeros con al menos 1 familiar
con_familiar = (combined['Familiares'] > 0).sum()
print(f"\nd) Pasajeros que viajaron con al menos un familiar: {con_familiar}")

# e) Edad más joven y más vieja
print(f"\ne) Edad más joven: {combined['Age'].min()} años  |  Edad más vieja: {combined['Age'].max()} años")

# f) Puerto de embarque
print("\nf) Pasajeros por puerto de embarque:")
print(combined['Embarked'].value_counts().rename({'S':'Southampton','C':'Cherburgo','Q':'Queenstown'}))

# g) Solos vs con familiares
solos = (combined['Familiares'] == 0).sum()
con_fam = (combined['Familiares'] > 0).sum()
print(f"\ng) Viajaron solos: {solos}  |  Con familiares: {con_fam}")
```

```python
print("\n" + "=" * 60)
print("9. ANÁLISIS DE SUPERVIVENCIA")
print("=" * 60)

# a) Mujeres vs hombres
print("\na) ¿Sobrevivieron más mujeres que hombres?")
sex_surv = train.groupby('Sex')['Survived'].agg(['sum','count'])
sex_surv['tasa'] = (sex_surv['sum'] / sex_surv['count'] * 100).round(2)
sex_surv.columns = ['Sobrevivieron','Total','Tasa (%)']
print(sex_surv)
print("  → SÍ. La tasa de supervivencia femenina (~74%) supera ampliamente la masculina (~19%).")

# b) Niños vs hombres adultos
print("\nb) ¿Sobrevivieron más niños que hombres adultos?")
train['grupo_etario'] = pd.cut(
    train['Age'],
    bins=[-1, 10, 17, 49, 120],
    labels=['Niño (<10)', 'Adolescente', 'Adulto (18-49)', 'Mayor (≥50)']
)
ninos  = train[train['grupo_etario'] == 'Niño (<10)']['Survived']
h_adu  = train[(train['grupo_etario'] == 'Adulto (18-49)') & (train['Sex'] == 'male')]['Survived']
print(f"  Tasa niños (<10): {ninos.mean()*100:.1f}%   |   Tasa hombres adultos: {h_adu.mean()*100:.1f}%")
print("  → SÍ. Los niños menores de 10 años sobrevivieron a mayor tasa que los hombres adultos.")

# c) Mayores de 50 o menores de 10 vs otros grupos
print("\nc) ¿Pasajeros >50 o <10 años sobrevivieron más que otros grupos?")
train['grupo_c'] = train['Age'].apply(
    lambda x: 'Extremos (<10 o >50)' if (x > 50 or x < 10) else 'Otros'
)
print(train.groupby('grupo_c')['Survived'].agg(['mean','sum','count'])
         .rename(columns={'mean':'Tasa','sum':'Sobrevivieron','count':'Total'})
         .assign(Tasa=lambda x: (x['Tasa']*100).round(2)))

# d) Southampton vs otros puertos
print("\nd) ¿Southampton sobrevivió más?")
emb_surv = train.groupby('Embarked')['Survived'].agg(['mean','sum','count'])
emb_surv.columns = ['Tasa','Sobrevivieron','Total']
emb_surv['Tasa (%)'] = (emb_surv['Tasa'] * 100).round(2)
print(emb_surv[['Sobrevivieron','Total','Tasa (%)']])
print("  → NO. Cherburgo (C) tiene la mayor tasa de supervivencia (~55%), seguido de Queenstown (~39%).")
print("    Southampton tiene la menor tasa (~34%), aunque embarcó la mayor cantidad de pasajeros.")

# e) Primera clase vs otras
print("\ne) ¿Primera clase sobrevivió más?")
cls_surv = train.groupby('Pclass')['Survived'].agg(['mean','sum','count'])
cls_surv.columns = ['Tasa','Sobrevivieron','Total']
cls_surv['Tasa (%)'] = (cls_surv['Tasa'] * 100).round(2)
print(cls_surv[['Sobrevivieron','Total','Tasa (%)']])
print("  → SÍ. Primera clase (~63%) sobrevivió a mayor tasa que segunda (~47%) y tercera (~24%).")
```

```python
print("\n" + "=" * 60)
print("10. GRUPOS FAMILIARES Y DISTRIBUCIÓN DE CABINAS")
print("=" * 60)

combined['grupo_familiar'] = pd.cut(
    combined['Familiares'],
    bins=[-1, 0, 3, 100],
    labels=['Grupo 1: Sin familiar', 'Grupo 2: Familia pequeña (1-3)', 'Grupo 3: Familia grande (≥4)']
)

print("\nDistribución de pasajeros por grupo familiar:")
print(combined['grupo_familiar'].value_counts().sort_index())

print("\nDistribución de cabinas por grupo familiar:")
combined['tiene_cabina'] = combined['Cabin'].notna().map({True:'Con cabina', False:'Sin cabina'})
tabla_cabina = pd.crosstab(combined['grupo_familiar'], combined['tiene_cabina'])
tabla_cabina['% Con cabina'] = (tabla_cabina['Con cabina'] / tabla_cabina.sum(axis=1) * 100).round(1)
print(tabla_cabina)

print("""
Interpretación: Las familias pequeñas (Grupo 2) y las personas que viajan
solas (Grupo 1) muestran patrones similares de cabina. El porcentaje de
pasajeros con cabina asignada es bajo en todos los grupos, lo que refleja
que la mayoría viajaba en tercera clase sin cabina individual.
""")
```

```python
print("=" * 60)
print("11. PROPORCIÓN DE SUPERVIVENCIA POR SEXO, GRUPO ETARIO Y CABINA")
print("=" * 60)

# Rehacer grupo_etario sobre train con columnas correctas
train['grupo_etario'] = pd.cut(
    train['Age'],
    bins=[-1, 9.99, 17.99, 49.99, 120],
    labels=['Niños (<10)', 'Adolescentes (10-17)', 'Adultos (18-49)', 'Mayores (≥50)']
)
train['tiene_cabina'] = train['Cabin'].notna().map({True:'Con cabina', False:'Sin cabina'})


tabla_11 = (
    train.groupby(['Sex', 'grupo_etario', 'tiene_cabina'], observed=True)['Survived']
    .agg(['mean', 'sum', 'count'])
    .rename(columns={'mean':'Tasa_superv','sum':'Sobrevivieron','count':'Total'})
)
tabla_11['Tasa (%)'] = (tabla_11['Tasa_superv'] * 100).round(1)
print(tabla_11[['Sobrevivieron','Total','Tasa (%)']].to_string())

print("""Conclusión: Las mujeres con cabina y los niños tienen las mayores tasas
de supervivencia, independientemente del grupo etario. Los hombres adultos
sin cabina (tercera clase) tienen las tasas más bajas. Esto refleja la
política "mujeres y niños primero" y la desigualdad de acceso a botes
según clase social.""")
```
